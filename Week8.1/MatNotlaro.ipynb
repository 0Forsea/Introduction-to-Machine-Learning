```python
import pandas as pd
```

```python
veri=pd.read_csv('student-mat.csv',sep=';')
```

```python
veri=veri[['G1','G2','G3','studytime','failures','absences','age']]
```

```python
veri.head(100)
```

```python
veri.rename(columns={'G1':'Not1',
                    'G2':'Not2',
                    'G3':'Final',},inplace=True)
veri.head()
```

```python
import numpy as np
y=np.array(veri['Final'])
X=np.array(veri.drop('Final',axis=1))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
veri.head()
```

```python
yeni_veri=np.array([[10,15,0,0,15,18]])
lr.predict(yeni_veri)
```
